```python
def Evaluation(cm):
    print(cm)
    TP = cm[0][0]
    FN = cm[0][1]
    FP = cm[1][0]
    TN = cm[1][1]
    Sensitivity = TP/(TP + FN)  #accuracy of positive class or Recall
    print("Sensitivity",Sensitivity*100)
    Specificity = TN/(TN + FP)  #accuracy of positive class or Recall
    print("Specificity",Specificity*100)
    Precision = TP /(TP + FP) # out of all total positive decisions how many are correct
    print("Precision",Precision*100)
    F_Score = 2 * (Precision*Sensitivity)/(Precision + Sensitivity)
    print('F_Score',F_Score)
    Accuracy = (TP + TN)/(TP + FN + FP + TN) # out of all positive and negative classes predicted accuratly
    print('Accuracy',Accuracy*100)
    Balanced_Accuracy = (Sensitivity + Specificity)/ 2
    print('Balanced_Accuracy', Balanced_Accuracy*100)
```

```python
X_train = df.iloc[:,:-1]
y_train = df.iloc[:,-1:]
```

```python
from sklearn.tree import DecisionTreeClassifier

# Building decision tree with entropy
model = DecisionTreeClassifier(criterion='entropy', max_depth=5, random_state=42)
```

```python
#Training the model
h = model.fit(X_train,y_train)
#predicting on the trained data
Train_cap = h.predict(X_train)
cm_train = confusion_matrix(y_train,Train_cap)
print("Train Dataset Metrics")
Evaluation(cm_train)
```

```python
#for train and test

X = df.drop('Risk Category', axis=1)  # Features (inputs)
y = df['Risk Category']               # Target (output)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,   # 20% test, 80% train (commonly used)
    random_state=42  # for reproducibility
)
```

```python
model1=DecisionTreeClassifier(criterion="entropy",max_depth=5,random_state=42)
```

```python
h = model1.fit(X_train,y_train)
Train_cap = h.predict(X_train)
cm_train = confusion_matrix(y_train,Train_cap)
print("Train Dataset Metrics")
Evaluation(cm_train)
Test_cap = h.predict(X_test)
cm_test = confusion_matrix(y_test,Test_cap)
print("Test Dataset Metrics")
Evaluation(cm_test)
```
